```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, feature, exposure
```

```python
# Загрузим и отобразим изображение
image = io.imread("photo_.jpg")
image_gray = color.rgb2gray(image)

plt.imshow(image_gray, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# 1. Вычисление градиентов
gradient_y, gradient_x = np.gradient(image_gray)
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)
gradient_angle = np.arctan2(gradient_y, gradient_x) * (180/np.pi)

plt.imshow(gradient_magnitude, cmap='magma')
plt.title('Gradient Magnitude')
plt.axis('off')
plt.show()
```

```python
plt.imshow(gradient_angle, cmap='hsv')
plt.title('Gradient Angle')
plt.axis('off')
plt.show()
```

```python
# 2. HOG Feature Extraction
fd, hog_image = feature.hog(image_gray, orientations=9, pixels_per_cell=(8, 8), 
                            cells_per_block=(2, 2), visualize=True, block_norm='L2-Hys')

plt.imshow(1-hog_image/np.max(hog_image))
plt.title('HOG Image')
plt.axis('off')
plt.show()
```
