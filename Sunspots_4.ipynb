```python
import pandas as pd
import numpy as np

def prepare_data(df, lags=10, diff=10, count_autocor=10):
    # Извлечение 10 лагов
    for i in range(1, lags+1):
        df[f'Lag_{i}'] = df['Sunspots'].shift(i)

    # Извлечение 10 разностей
    for i in range(1, diff+1):
        df[f'Diff_{i}'] = df['Sunspots'].diff(i)

    # Извлечение 10 автокорреляций
    for i in range(1, count_autocor+1):
        df[f'Autocorr_{i}'] = df['Sunspots'].autocorr(lag=i)

    # Удаление строк с NaN значениями
    return df
data_n = prepare_data(data)
```

```python
from sklearn.model_selection import TimeSeriesSplit

data_n = data_n.dropna()

# Разбиение данных на тренировочный и тестовый наборы
split_ratio = 0.8  # Процент данных для тренировки
split_index = int(split_ratio * len(data_n))

train_data = data_n.iloc[:split_index]
test_data = data_n.iloc[split_index:]

# Инициализация и применение RobustScaler к всем признакам, кроме целевой переменной
#scaler = RobustScaler()
scaler = StandardScaler()
train_data.iloc[:, 1:-1] = scaler.fit_transform(train_data.iloc[:, 1:-1])
test_data.iloc[:, 1:-1] = scaler.transform(test_data.iloc[:, 1:-1])

# Разделение на признаки и целевую переменную
X_train = train_data.iloc[:, 1:-1]
y_train = train_data['Target']
X_test = test_data.iloc[:, 1:-1]
y_test = test_data['Target']


```

```python
# Проверка результатов
print("X_train (нормализованные признаки для тренировки):\n", X_train)
print("\nX_test (нормализованные признаки для теста):\n", X_test)
print("\ny_train (целевая переменная для тренировки):\n", y_train)
print("\ny_test (целевая переменная для теста):\n", y_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping


# Создание модели Bidirectional RNN
model = Sequential()
model.add(Bidirectional(LSTM(64, return_sequences=False), input_shape=(X_train.shape[1], 1)))
model.add(Dense(1))

# Компиляция модели
model.compile(optimizer='adam', loss='mean_squared_error')

# Обучение модели
early_stopping = EarlyStopping(patience=5, restore_best_weights=True)  # Остановка обучения при отсутствии улучшений
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))  # Изменение формы данных для LSTM
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), callbacks=[early_stopping])

# Оценка модели на тестовых данных
loss = model.evaluate(X_test, y_test)
print(f"Loss on test data: {loss}")

```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

# Получите прогнозы модели на тестовых данных (предположим, что модель называется model)
y_pred = model.predict(X_test).ravel()

# MAE (Mean Absolute Error)
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (MAE): {mae}")

# MAPE (Mean Absolute Percentage Error)
def calculate_mape(y_true, y_pred):
    y_pred=y_pred.ravel()
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mape = calculate_mape(y_test, y_pred)
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
temp_df = pd.DataFrame(y_test.values,columns=['y_test'], index = y_test.index)
temp_df["y_pred"] = y_pred
```

Для реализации механизма внимания Bahdanau для временных рядов, мы можем использовать библиотеку TensorFlow и Keras. В данном случае, мы рассмотрим использование механизма внимания Bahdanau в контексте прогнозирования временных рядов.

Вам потребуется предварительно обработать данные временного ряда и подготовить их в виде входной последовательности временных шагов и соответствующих целевых значений. Ниже приведен код для реализации механизма внимания Bahdanau для временных рядов:


```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input, LSTM, Dense, Attention
from tensorflow.keras.models import Model

# Гиперпараметры
sequence_length = 30  # Длина входной последовательности временных шагов
input_dim = 1  # Размерность признаков в каждом временном шаге
hidden_units = 256  # Размерность скрытого слоя LSTM
output_dim = 1  # Размерность выходного слоя
attention_units = 64  # Размерность внимания Bahdanau

# Создание входных данных
input_sequence = Input(shape=(sequence_length, input_dim))

# LSTM слой
lstm_layer = LSTM(hidden_units, return_sequences=False)(input_sequence)

# lstm_layer2 = LSTM(hidden_units, return_sequences=False)(lstm_layer)
# Механизм внимания Bahdanau
attention_layer = Attention(use_scale=True)([lstm_layer, lstm_layer])

# Конкатенация LSTM выхода и слоя внимания
merged_layer = tf.concat([lstm_layer, attention_layer], axis=-1)

# Полносвязанный слой для прогнозирования
output_layer = Dense(output_dim)(merged_layer)

# Создание модели
model = Model(inputs=input_sequence, outputs=output_layer)

# Компиляция модели
model.compile(optimizer='adam', loss='mean_squared_error')

# Вывод структуры модели
model.summary()
```

```python
# Обучение модели
early_stopping = EarlyStopping(patience=5, restore_best_weights=True)  # Остановка обучения при отсутствии улучшений
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))  # Изменение формы данных для LSTM
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
model.fit(X_train, y_train, epochs=50, batch_size=128, validation_data=(X_test, y_test), callbacks=[early_stopping])

# Оценка модели на тестовых данных
loss = model.evaluate(X_test, y_test)
print(f"Loss on test data: {loss}")
```

```python
# Получите прогнозы модели на тестовых данных (предположим, что модель называется model)
y_pred = model.predict(X_test)

# Преобразуйте прогнозы в одномерный массив
y_pred = y_pred.ravel()

# MAE (Mean Absolute Error)
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (MAE): {mae}")

# MAPE (Mean Absolute Percentage Error)
def calculate_mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mape = calculate_mape(y_test, y_pred)
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")

# Создайте DataFrame для сравнения и визуализации результатов
temp_df = pd.DataFrame({'y_test': y_test, 'y_pred': y_pred}, index=y_test.index)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, LSTM, Dense, Attention
from tensorflow.keras.models import Model

# Гиперпараметры
sequence_length = 30  # Длина входной последовательности временных шагов
input_dim = 1  # Размерность признаков в каждом временном шаге
output_dim = 1  # Размерность выходного слоя
attention_units = 128  # Размерность внимания Luong

# Создание входных данных
input_sequence = Input(shape=(sequence_length, input_dim))

# LSTM слой
lstm_layer = LSTM(attention_units, return_sequences=True)(input_sequence)

# Механизм внимания Luong
attention_layer = Attention()([lstm_layer, lstm_layer])

# Взвешенная сумма выходов LSTM с использованием весов внимания
context_layer = tf.matmul(attention_layer, lstm_layer, transpose_a=True)

# Полносвязанный слой для прогнозирования
output_layer = Dense(output_dim)(context_layer)  # Нет необходимости объединять с LSTM

# Создание модели
model = Model(inputs=input_sequence, outputs=output_layer)

# Компиляция модели
model.compile(optimizer='adam', loss='mean_squared_error')

# Вывод структуры модели
model.summary()
```

```python
# Обучение модели
early_stopping = EarlyStopping(patience=5, restore_best_weights=True)  # Остановка обучения при отсутствии улучшений
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))  # Изменение формы данных для LSTM
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
model.fit(X_train, y_train, epochs=50, batch_size=128, validation_data=(X_test, y_test), callbacks=[early_stopping])

# Оценка модели на тестовых данных
loss = model.evaluate(X_test, y_test)
print(f"Loss on test data: {loss}")
```

```python
# Получите прогнозы модели на тестовых данных (предположим, что модель называется model)
y_pred = model.predict(X_test)

# Преобразуйте прогнозы в одномерный массив
# y_pred = y_pred.ravel()
y_pred = y_pred[:,-1,-1]
# MAE (Mean Absolute Error)
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (MAE): {mae}")

# MAPE (Mean Absolute Percentage Error)
def calculate_mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mape = calculate_mape(y_test, y_pred)
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")

# Создайте DataFrame для сравнения и визуализации результатов
temp_df = pd.DataFrame({'y_test': y_test, 'y_pred': y_pred}, index=y_test.index)
```
