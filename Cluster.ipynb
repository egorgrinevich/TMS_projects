```python
# Импорт необходимых библиотек и модулей
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN, KMeans

# 1. Изучение данных
# Загрузка набора данных
X, y = make_circles(n_samples=1000, factor=0.3, noise=0.1)
# Визуализация данных
plt.scatter(X[:, 0], X[:, 1], c='blue', marker='o', edgecolor='black', s=50)
plt.show()
```

```python
# значение esp<0.1 не позволяет определить центроид у внешнего круга
dbscan = DBSCAN(eps=0.05, min_samples=10)
clusters = dbscan.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap="viridis", marker='o', edgecolor='black', s=50)
plt.title('DBSCAN with eps=0.5, min_samples=5')
plt.show()
```

```python
# на этих данных комбинация максимального расстояния до соседа и количество точек вокруг ядра позволяет разделить на 2 класса
dbscan = DBSCAN(eps=0.21, min_samples=20)
clusters = dbscan.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap="viridis", marker='o', edgecolor='black', s=50)
plt.title('DBSCAN with eps=0.5, min_samples=5')
plt.show()
```

```python
# изменение eps. можно сказать, сто данные сформированны достаточно кучно. разделить на 2 класса можно многогими комбинациями eps и min_samples. 
# при этом если уменьшаем eps, то для желаемого результата нужно уменьшать min_samples. похоже на рассматривание под лупой, масштабирование
dbscan = DBSCAN(eps=0.15, min_samples=5)
clusters = dbscan.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap="viridis", marker='o', edgecolor='black', s=50)
plt.title('DBSCAN with eps=0.5, min_samples=5')
plt.show()
```

```python
# KMeans кластеризует, разделяя плоскость на части. 
kmeans = KMeans(n_clusters=2,  init='k-means++', n_init='warn', max_iter=1000, tol=0.01, verbose=0, random_state=None, copy_x=True, algorithm='lloyd')
clusters = kmeans.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap="viridis", marker='o', edgecolor='black', s=50)
plt.title('KMeans Clustering')
plt.show()
```
